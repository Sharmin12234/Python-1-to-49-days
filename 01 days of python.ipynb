# local vs global

```python
x = 1000 #global
def fun1():
    print('my num is :', x)
fun1()
```

```python
x = 1000 #global
def fun1():
    x = 500 #local
    print('my num is :', x)
fun1()
    
```

```python
x = 1000 #global
def fun1():
    x = 500 #local
    print('my num is :', x)
fun1()
print('my num is :', x)
```
