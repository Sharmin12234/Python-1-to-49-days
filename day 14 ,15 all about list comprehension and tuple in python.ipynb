# List comprehension

# Basic Syntax
[expression for item in iterable if condition]

# Simple List Comprehension

```python
squares = [x**2 for x  in range(10)]
print(squares)
```

```python
[x for x  in range(10)]
```

```python
 for x  in range(10):
        print(x)
```

# List Comprehension with condition

```python
evens = [x for x in range(10) if x % 2 == 0]
print(evens)
```

```python
data = ' I love data science'
[ x  for x in  data]
```

```python
data = ' I love data science'
[ x  for x in  data if x!='I']
```

```python
data = ' I love data science'
data = data.split()
[ x  for x in  data if x!='I']
```

# creating a matrix using list comprehension in python

```python
matrix = [[j for j in range (5)] for i in range(3)]
print(matrix)
```

# function using in list comprehension

```python
words = ['python','list','comprehension','example']
lengths = [len(words) for words in words]
print(lengths)
```
