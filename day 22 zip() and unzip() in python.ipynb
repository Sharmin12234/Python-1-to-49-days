# zip()

```python
name = [ 'sharmin','subha','maisa']
varsity = ['diu','aiub','diu']
```

```python
#('sharmin','diu'), ('subha','aiub',......)
zipped = list(zip(name, varsity))
print(zipped)
```

```python
zipped = tuple(zip(name, varsity))
print(zipped)
```

```python
zipped = dict(zip(name, varsity))
print(zipped)
```

```python
zipped.keys()
```

```python
zipped.values()
```

```python
zipped = set(zip(name, varsity))
print(zipped)
```

```python
zipped = frozenset(zip(name, varsity))
print(zipped)
```

# unzip

```python
zipped = list(zip(name, varsity))
print(zipped)
```

```python
name,varsity = zip(*zipped)
```
