# search and replace

```python
search_world = 'Bercelona'
replace_world ='Bayeran Munich'
with open(r'text1.docx','r') as file:
    data = file.read()
    data =data.replace(search_world,replace_world)
```

```python
with open(r'text1.docx','w') as file:
    file.write(data)
```
