# unpacking/unzipp

```python
ids = (1,2,3)
name = ('subha','sadiya','tanjina')
```

```python
zipped = tuple(zip(ids,name))
zipped
```

```python
ids, name = zip(*zipped)
```

```python
name
```

```python
zipped[0]
```

```python
ids, name = zipped[0]
```

```python
#unpackin using loop
for ids, names in zipped:
    print(ids, names)
```

```python
#enumerating: Iterate over pairs
for data in enumerate(zipped):
    #upack pair: ids name
    ids, name = data
    print(data)
```

```python
#enumerating: Iterate over pairs
for index,pair in enumerate(zipped):
    #upack pair: ids name
    ids, name = pair
    print('rank {},id {},name {}'.format(index,ids,name))
```
