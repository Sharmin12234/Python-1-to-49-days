# break statement

```python
for i in range(100):
        print(i)
    
```

```python
for i in range(100):
    if i == 50:
        break
    print(i)
        
```

```python
for i in range(1,100):
    if i == 11:
        break
    print(i)
        
```

```python
for i in range(1,100):
    if i == 11:
        break
    print('3*',i,'=',3*i)
        
```

```python
i = 1
while i<=100:
    if i == 11:
        break
    print('5*',i,'=',5*i)
    i = i+1
        
```
