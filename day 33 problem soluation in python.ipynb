# problem 2: list reverseal
create a function that reverses a given list and return the reversed list 

original_list=[10,20,30,40,50]

reversed list[50,40,30,20,10]

```python
def revlist(original_list):
    print(f'rev i {original_list}')
    
    result = reversed(original_list)
    return result

original_list=[10,20,30,40,50]
revlist(original_list)
    
```
