# paramiter vs argument

```python
def add_num(x,y):
    result = x+y
    return result
add_num(10,20)
```

```python
def add_num(x,y):
    result = x
    return result
add_num(10,20)
```

```python
1.positional argument
2.keyword argument
3.default argument
4.variable length argument
```

# 1.positional argument

```python
def add_num(x,y):
    result = x+y
    return result
add_num(10,20)
```

```python
def add_num(x,y):
    result = x
    return result
add_num(10,20)
```

# 2.keyword argument

```python
def per_inf(job, income):
    print('job = ',job)
    print('income = ',income)
per_inf(income=20000,job='data science')
```

# 3.default argument

```python
def per_inf(job, income = 20000):
    print('job = ',job)
    print('income = ',income)
per_inf('data science')
```

```python
def per_inf(job, income = 20000):
    print('job = ',job)
    print('income = ',income)
per_inf('data science',10000)
```

# 3.variable length argument

```python
def var_len(*x):
    print(x)
    result = sum(x)
    return result
    
var_len(10,20,30,40,50)
```

```python
def var_len(x,*y):
    print(x)
    print(y)
    
var_len(10,20,30,40,50)
```
