# lambda anonymous function

# function

```python
def sum(x,y):
    return x+y
sum(4,5)
```

lambda anonymous function

```python
f = lambda x,y:x+y
f(4,5)
```

```python
f = lambda x,y:x*y
f(4,5)
```

```python
f = lambda x,y:x/y
f(4,5)
```

# lambda in higher order


```python
def op_h(x,y,operator):
    return operator(x,y)
op_h(4,5,lambda x,y:x+y)
```
