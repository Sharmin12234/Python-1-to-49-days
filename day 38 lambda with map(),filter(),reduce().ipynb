# lambda with map(),filter(),reduce()

```python
num = [1,2,4,5,7,10]
num
```

```python
def sq(x):
    return x**2
sq(3)
```

```python
num
```

```python
# map(function,iteable object)
num = list(map(lambda x:x**2,num))
```

```python
num
```

```python
num
```

```python
num = list(filter(lambda num: num%2==1,num))
num
```

```python
from functools import reduce
num = reduce(lambda a,b:a+b,num)
```
