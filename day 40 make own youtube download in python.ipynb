```python
from pytube import Youtube
```

```python
ur! = youtube video link dite hobe
```

```python
yt = Youtube(ur!)
```

```python
stream = yt.stream.get_highest_resolution()
```

```python
stream.download()
```

```python
stream = yt.stream.filter(progressive=True, res='720p').first()
if stream:
    stream.download()
    print('Download Completed')
else:
    print('720p not available')
```
