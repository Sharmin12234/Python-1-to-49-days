# using generator function

```python
def count_up(max):
    count = 1
    while count <= max:
        yield count
        count += 1
counter = count_up(5)
```

```python
count_up(max)
```

# list

```python
def evens_numbers_list(limit):
    numbers = []
    num = 0
    while num < limit:
        numbers.append(num)
        num += 2
    return numbers
    
          
```

```python
evens = evens_numbers_list(10)
evens
```

```python
data_list = [10,20,30,40,50]

def data_generator(data):
    for item in data:
        yield item
        
gen = data_generator(data_list)

for item in gen:
    print(item)
```

```python
def evens_numbers():
    num = 0
    while True:
        yield num
        num += 2
        
gen = evens_numbers()

for _ in range(5):
    print(next(gen))
```

# list comprehension vs generator

```python
#list comprehension
ev_sq = [x**2 for x in range(10) if x%2== 0]
ev_sq
```

```python
sys.getsizeof(ev_sq)
```

```python
#generator
even_square_gen = (x**2 for x in range(10) if x % 2 == 0)
for square in even_square_gen:
    print(square)
```

```python
sys.getsizeof(even_square_gen)
```

```python
even_square_gen = tuple(x**2 for x in range(10) if x % 2 == 0)
for square in even_square_gen:
    print(square)
```

```python
even_square_gen = list(x**2 for x in range(10) if x % 2 == 0)
for square in even_square_gen:
    print(square)
```
